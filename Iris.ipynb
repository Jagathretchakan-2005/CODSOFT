```python
X = df.drop("species", axis=1) #Features
y = df['species'] #Target Variable
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test) #Predictions of the test data 
```

```python
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_pred: {y_pred.shape}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test['sepal_length'], y_test, color='blue', label='Actual data')
plt.scatter(X_test['sepal_length'], y_pred, color='red', label='Predicted data', alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Species')
plt.title('Comparison of Actual vs Predicted for Sepal Length')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, classification_report

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Print the classification report
report = classification_report(y_test, y_pred)
print(f"Classification Report:\n{report}")
```
